```python
from tkinter import *
from tkinter import messagebox


# ADD FUNCTION TO ADD CONACTS
def add():
    ne = name_entry.get().strip()
    pe = phone_entry.get().strip()
    if not ne or not pe:
        messagebox.showwarning('An Error Occurred !!!', 'Both Name and Phone must be entered')
    elif not pe.isdigit():
        messagebox.showwarning('Invalid Phone Number', 'Phone number should contain only digits')
    elif len(pe) != 10:
        messagebox.showwarning('Invalid Phone Number', 'Phone number must be exactly 10 digits')
    else:
        entry_text = f"Name: {ne} | Phone: {pe}"
        tasklist_box.insert(END, entry_text)
        name_entry.delete(0, END)
        phone_entry.delete(0, END)


# MARK FUNCTION TO MARK ANY CONTACT
def mark():
    index = tasklist_box.curselection()
    if index:
        task_text = tasklist_box.get(index)
        if not task_text.startswith("✔"):  # Prevent double marking
            tasklist_box.delete(index)
            tasklist_box.insert(index, f"✔ {task_text}")
    else:
        messagebox.showwarning('An Error Occured !!!','Select any task to mark')



# UPDATE FUNCTION
def update():
    selected = tasklist_box.curselection()
    new_name = name_entry.get().strip()
    new_phone = phone_entry.get().strip()

    if not selected:
        messagebox.showwarning("Update Error", "Select a contact to update")
    elif not new_name or not new_phone:
        messagebox.showwarning("Update Error", "Both Name and Phone must be entered")
    elif not new_phone.isdigit():
        messagebox.showwarning("Invalid Phone Number", "Phone number should contain only digits")
    elif len(new_phone) != 10:
        messagebox.showwarning("Invalid Phone Number", "Phone number must be exactly 10 digits")
    else:
        updated_contact = f"Name: {new_name} | Phone: {new_phone}"
        tasklist_box.delete(selected)
        tasklist_box.insert(selected, updated_contact)
        name_entry.delete(0, END)
        phone_entry.delete(0, END)



# SEARCH FUNCTION
def search():
    name_to_search = name_entry.get().strip().lower()
    if not name_to_search:
        messagebox.showwarning("Search Error", "Enter a name to search")
        return

    found = False
    for i in range(tasklist_box.size()):
        item = tasklist_box.get(i).lower()
        if name_to_search in item:
            tasklist_box.selection_clear(0, END)
            tasklist_box.selection_set(i)
            tasklist_box.see(i)
            found = True
            break

    if not found:
        messagebox.showinfo("Search Result", "No matching contact found")



# DELETE FUNCTION
def delete():
    select_task_index = tasklist_box.curselection()
    if select_task_index:
        tasklist_box.delete(select_task_index)
    else:
        messagebox.showwarning('An Error Occured !!!','Select task to remove')


# CLEAR ALL CONTACTS
def clear():
    tasklist_box.delete(0, END)





# CREATE OBJECT
root = Tk()
root.title("Contact Book")
root.configure(bg="#f0edff")


# HEADING
l = Label(root, text="Contact Book", bg="#f0edff", font=("Georgia", 20, "bold"), fg="#043d01", padx=20, pady=30)
l.pack()


# NAME AND PHONE FRAME
frame = Frame(root, bg="#f0edff")
frame.pack(pady=10)

# NAME LABEL
name = Label(frame, text="NAME", bg="#f0edff", font=("arial", 10, "bold"), anchor="w")
name.grid(row=0, column=0, padx=5, pady=2, sticky="w")

# NAME ENTRY
name_entry = Entry(frame, width=40, font=("arial", 12))
name_entry.grid(row=1, column=0, padx=5, pady=5)

# PHONE LABEL
phone = Label(frame, text="PHONE No.", bg="#f0edff", font=("arial", 10, "bold"), anchor="w")
phone.grid(row=2, column=0, padx=5, pady=2, sticky="w")

# PHONE ENTRY
phone_entry = Entry(frame, width=40, font=("arial", 12))
phone_entry.grid(row=3, column=0, padx=5, pady=5)


# ADD BUTTON
add_b = Button(frame, text=" ADD contact ", command=add, font=("arial", 10, "bold"))
add_b.grid(row=3, column=1, padx=10, pady=5, sticky="n")


# LIST BOX 
task_frame = Frame(frame, bg="#f0edff")
task_frame.grid(row=5, column=0, padx=5, pady=5)

tasklist_box = Listbox(task_frame, width=40, height=15, font=("arial", 12), bg="#f7f7f2")
tasklist_box.grid(row=0, column=0, padx=10, pady=5)


# BUTTON FRAME
button_frame = Frame(frame, bg="#f0edff")
button_frame.grid(row=5, column=1, padx=10, pady=5, sticky="n")

m_b = Button(button_frame, text="MARK", command=mark, font=("arial", 10, "bold"))
m_b.pack(pady=5, fill="x")

u_b = Button(button_frame, text="UPDATE", command=update, font=("arial", 10, "bold"))
u_b.pack(pady=5, fill="x")

s_b = Button(button_frame, text="SEARCH", command=search, font=("arial", 10, "bold"))
s_b.pack(pady=5, fill="x")

r_b = Button(button_frame, text="REMOVE", command=delete, font=("arial", 10, "bold"))
r_b.pack(pady=5, fill="x")

c_b = Button(button_frame, text="CLEAR ALL", command=clear, font=("arial", 10, "bold"))
c_b.pack(pady=5, fill="x")



# EXECUTE TKINTER
root.mainloop()
```
